```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
URL = "https://raw.githubusercontent.com/yash-1994/ML-learning-fcc-exmple-MAGIC-ML-learning-fcc-exmple/main/magic%2Bgamma%2Btelescope/magic04.data"
data = pd.read_csv(URL,header=None)
data.head(5)
```

```python
data.columns = ['fLength', 'fWidth', 'fSize', 'fConc', 'fConc1', 'fAsym', 'fM3Long', 'fM3Trans', 'fAlpha', 'fDist', 'class']
data["class"].unique()
data["class"] = (data["class"] == "g").astype(int)
```

```python
for label in data.columns:
  plt.hist(data[data["class"] == 1][label],color='blue',label='gamma',alpha=0.7)
  plt.hist(data[data["class"] == 0][label],color='red',label='hadron',alpha=0.7)
  plt.title(label)
  plt.ylabel("Probability")
  plt.xlabel(label)
  plt.legend()
  plt.show()
```
